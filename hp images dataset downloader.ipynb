```python
import requests
import json
import uuid
import requests
import pathlib
import os
import hashlib
import matplotlib.pyplot as plt
import numpy as np
import IPython
import cv2
import uuid

def get_adult_image_link(url, file="deploy-adult.txt"):

    response = requests.get(url)
    links = json.loads(response.content)
    urls = []
    for link in links:
        url = str(link['t_url'])
        if not url == "http://seethisinaction.com/servlet/files/12":
            with open(file, 'a') as f:
                f.write(url+"\n")

            urls.append(url)
        
    print(urls)


def get_neutal_images_url(category, page_limit=2, file="deploy-rnf.txt"):
    !rm $file
    base_url = "https://stocksnap.io/api/search-photos/{0}/relevance/desc/{1}"
    for i in range(1, page_limit):
        url = base_url.format(category, i)
        print(category, " : ", i, " / ", page_limit, ' => ', url)
        response = requests.get(url)
        links = json.loads(response.content)["results"]
        image_url = "https://cdn.stocksnap.io/img-thumbs/280h/{}.jpg"
        for link in links:   
            url = image_url.format(link['img_id'])
            with open(file, 'a') as f:
                f.write(url+"\n")
                


def preview_from_txt_urls(src='deploy-rnf.txt', start=0, limit=10, figsize=(30, 30), verbose=False, dimensions=(224, 224)):
    urls = []
    
    with open(src) as f:
        lines = [line.rstrip() for line in f]
        tot = len(lines)
        print(tot, " urls found")
        count = 0
        for url in lines[start:limit]:
            count+=1
            urls.append(url)
            try:
                if verbose:
                    print(count, "/", tot, "dwd => ", url)
                
            except Exception as wrong:
                if verbose:
                    print(count, "/", tot, "error => ",wrong)
                pass
    
    preview_from_urls(urls, figsize=figsize, verbose=verbose, dimensions=dimensions)
        
        
def preview_from_urls(urls, figsize=(30, 30), verbose=False, dimensions=(224, 224)):
    images = []
    tot = len(urls)
    count=0
    for url in urls:
            count+=1
            try:
                if verbose:
                    print(count, "/", tot, "dwd => ", url)
                req = requests.get(url, stream=True)
                image = np.asarray(bytearray(req.content), dtype="uint8")
                imageBGR = cv2.imdecode(image, cv2.IMREAD_COLOR)
                imageRGB = cv2.cvtColor(imageBGR , cv2.COLOR_BGR2RGB)

                images.append(cv2.resize(imageRGB, dimensions, interpolation = cv2.INTER_AREA)/255)
            except Exception as wrong:
                if verbose:
                    print(count, "/", tot, "error => ",wrong)
                pass
   
    rangeTot = len(images)

    plt.figure(figsize=figsize)
    if len(images) == 1:
        plt.title("1")
        plt.imshow(images[0])
    else:  
        count = 0
        for i in range(rangeTot):
            count+=1
            plt.subplot(rangeTot,int((rangeTot)/2),i+1)
            plt.subplots_adjust(hspace = 0.8)
            #color = "blue" if predicted_ids[i] == label_batch[i] else "red"
            plt.title(str(count))#, color=color)
            #plt.imshow(images[i]/255 if predicted_labels[i]=="neutral" else ndimage.gaussian_filter(images[i]/255, sigma=2))
            plt.imshow(images[i])
        plt.show()
        
        


def download_images_from_txt_urls(src='deploy-rnf.txt', path="images_reinforce_test/neutral", start=0, limit=0, verbose=False):
    urls = []
    
    with open(src) as f:
        lines = [line.rstrip() for line in f]
        tot = len(lines)
        if limit==0:
            limit = tot
        print(tot, " urls found")
        count = 0
        for url in lines[start:limit]:
            count+=1
            urls.append(url)
            try:
                if verbose:
                    print(count, "/", tot, "dwd => ", url)
                
            except Exception as wrong:
                if verbose:
                    print(count, "/", tot, "error => ",wrong)
                pass
    
    download_images_from_urls(urls,path=path, verbose=verbose)
        
        
def download_images_from_urls(urls, path, verbose=False):
    ids = str(uuid.uuid1())
    tot = len(urls)
    count=0
    for url in urls:
        count+=1
        try:
            if verbose:
                print(count, "/", tot, "dwd => ", url)

            req = requests.get(url, stream=True, timeout=10)
            currentId = str(uuid.uuid1())
            file = path+"/"+ids+currentId+".jpg"
            
            with open(file, 'wb') as f:
                f.write(req.content)

        except Exception as wrong:
            if verbose:
                print(count, "/", tot, "error => ",wrong)
            pass
```
